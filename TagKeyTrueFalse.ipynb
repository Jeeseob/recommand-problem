```python
from pathlib import Path

content = Path("taglist.txt").read_text(encoding = 'UTF-8').replace('\n', ' ')
print(type(content))
```

```python
import re

pattern = r'"key": ".*?"'

taglist = re.findall(pattern, content)
```

```python
pattern1 = r'"key":'
pattern2 = r'\''
```

```python
taglist = str(taglist)
taglist = re.sub(pattern1, "", taglist)
taglist = re.sub(pattern2, "", taglist)
taglist = re.sub('\[', "", taglist)
taglist = re.sub('\]', "", taglist)
taglist = re.sub('\"', "", taglist)
```

```python
print(taglist)
```

```python
taglist = re.sub(" ", "", taglist)
```

```python
taglist = taglist.split(",")
```

```python
print(taglist)
```

```python
df['tag-key'] = df['tag-key'].str.replace(pat = r'[^\w]', repl = r' ', regex = True)
df['tag-key']
```

```python
count1 = 0
for _ in range(0, 191):
    df[taglist[count1]] = np.nan
    count1 = count1 + 1
```

```python
for i in taglist:
   df[i] = df['tag-key'].str.contains(i, case = False)
```
